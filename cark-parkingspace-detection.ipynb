```python
import cv2
import pandas as pd
from ultralytics import YOLO
import os
import matplotlib.pyplot as plt # Import matplotlib.pyplot


# Load trained YOLOv8 model (custom or pretrained)
model = YOLO('/content/runs/detect/train/weights/last.pt')  # Replace with your trained model, e.g., 'best.pt'

# Read the input image
img_path = '/content/drive/MyDrive/yolov8_parking_dataset/test/images/2013-03-05_07_25_01_jpg.rf.f6d0b389c2bbfe0862d7e70ad92000d4.jpg'  # Change this to your uploaded image path
#img_path='/content/drive/MyDrive/yolov8_parking_dataset/train/images/2012-09-12_12_48_19_jpg.rf.04ad2c8f45008da9ae99a3635e22190a.jpg'
image = cv2.imread(img_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# Inference
results = model.predict(source=img_path, conf=0.5)

# Parse detections
boxes = results[0].boxes
output_data = []
available_boxes = [box for box in boxes if int(box.cls[0]) == 0]  # Empty slots
occupied_boxes = [box for box in boxes if int(box.cls[0]) == 1]   # Cars
available_slots = len(available_boxes)
occupied_slots = len(occupied_boxes)
total_slots = available_slots + occupied_slots
slot_id = 1
for box in available_boxes:
  x1, y1, x2, y2 = map(int, box.xyxy[0])
  cv2.rectangle(image_rgb, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green for available
  cv2.putText(image_rgb, 'Available', (x1, y1 - 10),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

  output_data.append({
            'Slot ID': slot_id,
                    'X1': x1,
                            'Y1': y1,
                                    'X2': x2,
                                            'Y2': y2,
                                                    'Status': 'Available'
                                                        })

  slot_id += 1
for box in occupied_boxes:
  x1, y1, x2, y2 = map(int, box.xyxy[0])
  cv2.rectangle(image_rgb, (x1, y1), (x2, y2), (255, 0, 0), 2)  # Red for occupied
  cv2.putText(image_rgb, 'Occupied', (x1, y1 - 10),
                             cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
  output_data.append({
              'Slot ID': slot_id,
                      'X1': x1,
                              'Y1': y1,
                                      'X2': x2,
                                              'Y2': y2,
                                                      'Status': 'Occupied'
                                                          })
  slot_id += 1


df = pd.DataFrame(output_data)
df.to_csv('parking_status_output.csv', index=False)

# Save summary CSV
summary_df = pd.DataFrame({
    'Total Number of Slots': [total_slots],
        'Occupied Slots': [occupied_slots],
            'Available Slots': [available_slots]
            })
summary_df.to_csv('parking_summary.csv', index=False)
plt.figure(figsize=(12, 8))
plt.imshow(image_rgb)
plt.axis('off')
plt.title('Parking Slot Occupancy Detection')
plt.show()

# Print summary
print("=== Parking Summary ===")
print(f"Total Slots     : {total_slots}")
print(f"Occupied Slots  : {occupied_slots}")
print(f"Available Slots : {available_slots}")
```
